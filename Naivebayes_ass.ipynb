```python
import pandas as pd
import numpy as np
```

```python
sal_test = pd.read_csv(r"C:\Users\Aruna\Desktop\Assgnmnts-DS\Navbias\SalaryData_Test.csv")
sal_train = pd.read_csv(r"C:\Users\Aruna\Desktop\Assgnmnts-DS\Navbias\SalaryData_Train.csv")
```

```python
# Applying one hat encoding
sal_test_enc = pd.get_dummies(sal_test, columns=['workclass','education','maritalstatus','occupation','relationship','race','sex','native','Salary'], drop_first=True)
sal_train_enc = pd.get_dummies(sal_train, columns=['workclass','education','maritalstatus','occupation','relationship','race','sex','native','Salary'],drop_first=True)
```

Splitting data x_train, x_test and y_train,y_test

```python
# Splitting data x_train, x_test and y_train,y_test
x_train = sal_train_enc.drop("Salary_ >50K", axis=1)
x_test = sal_test_enc.drop("Salary_ >50K", axis=1)

y_train = sal_train_enc["Salary_ >50K"]
y_test = sal_test_enc["Salary_ >50K"]
```

```python
y_test
```

# Preparing Model

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
# intiating model
nb_classifier = GaussianNB()
```

```python
# fitting the model x_train,y_train
nb_classifier.fit(x_train,y_train)
```

```python
# Predicting the x_test data with fitted model
y_pred = nb_classifier.predict(x_test)
```

```python
# calculating the accuracy for test data
accuracy = accuracy_score(y_test,y_pred)
print("Accuracy Score:", accuracy*100)
```

```python
# Classification report
c_report = classification_report(y_test,y_pred)
print("Classification Report:\n", c_report)
```
