```python
book_df_pivot = book_df.pivot_table(index='User.ID',
                                 columns='Book.Title',
                                 values='Book.Rating',)
```

```python
book_df_pivot
```

```python
book_df_pivot.fillna(0, inplace=True)
```

```python
book_df_pivot
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
book_cos = 1 - pairwise_distances(book_df_pivot.values,metric='cosine')
```

```python
book_cos
```

```python
#Store the results in a dataframe
book_cos1 = pd.DataFrame(book_cos)
book_cos1
```

```python
#Set the index and column names to user ids 
book_cos1.index = book_df['User.ID'].unique()
book_cos1.columns = book_df['User.ID'].unique()
```

```python
book_cos1.iloc[0:10,0:10]
```

```python
np.fill_diagonal(book_cos, 0)
book_cos1.iloc[0:15, 0:15]
```

```python
# most similar users
book_cos1.idxmax(axis=1)[0:20]
```
